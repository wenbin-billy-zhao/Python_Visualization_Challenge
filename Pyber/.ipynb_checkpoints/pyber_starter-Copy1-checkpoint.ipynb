## Pyber Data Visualization
### the following graphs demonstrate the relationships between Uber drivers, total rides, average fare, city type

```python
## The following functions were decleared to manipulate color attributes for the plots
## Because I would like to use only one dataframe and one plot to generate the entire bubble chart, rather than using
## three sub-plots to overlap each other, I built the following functions to turn city types from string to numerical values
## therefore they can be aggragated by using Groupby function.
def set_color(a):
    if a == 'Rural':
        return 3
    elif a == 'Urban':
        return 2
    else:
        return 1

def set_color2(b):
    if b == 1:
        return 'lightskyblue'
    elif b == 2:
        return 'lightcoral'
    else:
        return 'gold'

## this function is used to generate custom legends for the scatter plot
def set_color3(c):
    if c == 'Urban':
        return 'lightcoral'
    elif c == 'Rural':
        return 'Gold'
    else:
        return 'lightskyblue'
```

```python
# Magic function
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
df = pd.merge(city_df, ride_df, on='city')

# add a column of numerical data representing types in the combined dataframe 
cList = list(df['type'])
cList = [set_color(x) for x in cList]
df['color'] = cList
df['type'].unique()
df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
xlim = df.groupby(['type','city'])['city'].count()
ylim = df.groupby(['type','city'])['fare'].mean()
slim = df.groupby(['type','city'])['driver_count'].min()
clim = df.groupby(['type','city'])['color'].min()

# join the series back into a single dataframe with two level indexes (type, city)
df2 = pd.DataFrame(
    {
        'Rides':xlim,
        'Avg Fare': ylim,
        'Driver Count': slim,
        'Color': clim
    }
)

# convert color code columns back to color names 
cList2 = list(df2['Color'])
cList2 = [set_color2(y) for y in cList2]

df2['color2'] = cList2
df2.head()
```

```python
# the fun begins: 
# build a scatter plot

PyberPlot1 = df2.plot.scatter('Rides','Avg Fare', c=df2['color2'], s=(df2['Driver Count']*10), alpha=0.8, edgecolors='black')

# Incorporate the other graph properties
plt.grid()
plt.title('Pyber Ride Sharing Data (2016)')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')

# build the custom legend
for myType in ['Urban', 'Suburban', 'Rural']:
    plt.scatter([], [], s=30, label=myType, c=set_color3(myType), edgecolors='black', alpha=0.8)
plt.legend(scatterpoints=1, frameon=True, title='City Types')

# build a custom text label on the side of the graph
textstr = 'Note:\nCircle size correlates with driver count per city.'
plt.text(0.92, 0.6, textstr, fontsize=11, transform=plt.gcf().transFigure)

# Save Figure
plt.savefig('bubble_plot.png')

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
totalFares = df.groupby(['type'])['fare'].sum().to_dict()
sizes = totalFares.values()
labels = totalFares.keys()
explode = (0, 0, 0.1)
colors = ['gold','lightskyblue','lightcoral']

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, shadow=True, startangle=150, autopct="%1.1f%%")
plt.axis('auto')
plt.title('% of Total Fares by City Type', fontsize=12)

# Save Figure
plt.savefig('total-fare-by-city.png')

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
totalRides = df.groupby(['type'])['ride_id'].count().to_dict()
sizes1 = totalRides.values()
labels1 = totalRides.keys()
explode = (0, 0, 0.1)
colors = ['gold','lightskyblue','lightcoral']

# build the pie chart
plt.pie(sizes1, explode=explode, labels=labels1, colors=colors, shadow=True, startangle=150, autopct="%1.1f%%")
plt.axis('auto')
plt.title('% of Total Rides by City Type', fontsize=12)

# Save Figure
plt.savefig('pct-of-rides-by-city.png')

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
totalDrivers = df2.groupby(['type'])['Driver Count'].sum().to_dict()

# build the pie chart
sizes2 = totalDrivers.values()
labels2 = totalDrivers.keys()
explode = (0, 0, 0.1)
colors = ['gold','lightskyblue','lightcoral']

plt.pie(sizes2, explode = explode, labels=labels2, colors=colors, shadow=True, startangle=155, autopct="%1.1f%%")
plt.axis('auto')
plt.title('% of Total Drivers by City Type', fontsize=12)

# Save Figure
plt.savefig('pct-of-drivers-by-city.png')

# Show Figure
plt.show()
```
